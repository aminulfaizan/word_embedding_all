```python
import time
from gensim.models.fasttext import FastText

start_time = time.time()

# your training code goes here
# Train a fastText model
model = FastText(token_corpus, 
                sg=1,
                hs=0,
                min_n=2,
                max_n=5,
                negative=5,
                vector_size=100,
                window=5,
                epochs = 20,
                min_count=25,
                workers=6)

end_time = time.time()

print("Training time:", end_time - start_time, "seconds")

model.save('gensim_fasttext_skip_dim100_win5_min25_epoch20.bin')    

#sg=0 mean use the CBOW
#hs = 0 softmax not using
#negative = 5 mean using negative sampling with value 5
#min_count =5 take only those words in training, occurring time least 5
#workers mean thread amount should use for this machine
#windows mean right and left side of the targeted word
# Save the model to disk
```
