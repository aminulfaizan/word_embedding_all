```python
import nltk
from nltk.stem import WordNetLemmatizer

# Download the Bengali lemmatizer data
nltk.download("indian")

# Initialize the Bengali lemmatizer
lemmatizer = WordNetLemmatizer()

# Define a function to lemmatize Bengali words
def bengali_lemmatizer(word):
    return lemmatizer.lemmatize(word, "bn")

# Example data with predictions and valid words
examples = [
    [("দিল্লিই", 0.790497362613678), ("ভারত", 0.7845825552940369), ("ভারতও", 0.7774765491485596)],
    [("কাতারপাকিস্তান", 0.6400704979896545), ("বাকিস্তান", 0.5991875529289246), ("তাজাকিস্তান", 0.5888345837593079)],
    [("ফিলিস্তিনের", 0.7425485253334045), ("ফিলিস্তিনই", 0.7252228856086731), ("ফিলিস্তিনইসরায়েল", 0.7249541878700256)]
]

# Process and filter predictions for all examples
for i, example in enumerate(examples, 1):
    filtered_predictions = [bengali_lemmatizer(word[0]) for word in example]
    print(f"Example {i}: {filtered_predictions}")
```
