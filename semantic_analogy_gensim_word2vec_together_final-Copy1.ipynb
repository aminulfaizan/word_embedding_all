```python
# Define ground truth
ground_truth = {
    ('ইরান', 'মস্কো', 'দোহা'): True,
    ('রাশিয়া', 'মস্কো', 'দোহা'): True,
}

# Define a similarity score threshold
similarity_threshold = 0.75  # Adjust this threshold as needed

# Initialize variables for true positives, false positives, and false negatives
true_positives = 0
false_positives = 0
false_negatives = 0

# Iterate through predicted analogies and their similarity scores
for analogy, similar_words_with_scores in predicted_results.items():
    # Extract the words from the analogy
    word1, word2, word3 = analogy
    
    # Find the most similar word and its similarity score
    most_similar_word, highest_similarity_score = max(similar_words_with_scores, key=lambda x: x[1])
    
    # Check if the highest similarity score is above the threshold
    if highest_similarity_score >= similarity_threshold:
        # Check if the most similar word matches any word in the ground truth analogy
        if most_similar_word in ground_truth.get((word1, word2, word3), []):
            true_positives += 1
        else:
            false_positives += 1
    else:
        # The predicted analogy was not considered a match
        false_negatives += 1

# Calculate precision, recall, and F1-score
precision = true_positives / (true_positives + false_positives)
recall = true_positives / (true_positives + false_negatives)
f1 = 2 * (precision * recall) / (precision + recall)

# Calculate accuracy
accuracy = true_positives / len(predicted_results)

# Print the results
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```
